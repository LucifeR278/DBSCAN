## Density-based spatial clustering of applications with noise (DBSCAN):-
it groups together points that are closely packed (points with many nearby neighbors), and marks as outliers points that lie alone in low-density regions (those whose nearest neighbors are too far away).




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,y=make_moons(n_samples=250,noise=0.05)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
## feature scaling(Standard Scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
dbscan=DBSCAN(eps=0.3)
dbscan.fit(X_scaled)
```

```python
y_pred=dbscan.labels_
```

```python
## Plot with scaled data
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=y_pred)
plt.show()
```

```python
#Plot with the unscaled data
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.show()
```
